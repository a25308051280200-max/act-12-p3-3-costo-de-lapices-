```python
# Programa para calcular el pago por la compra de lápices
# Si la cantidad es > 1000, el costo es 0.85
# De lo contrario, el costo es 0.90

# 1. Datos de entrada
X = int(input("Ingrese la cantidad de lápices a comprar: "))

# 2. Estructura condicional para determinar el pago
if X >= 1000:
    PAG = X * 0.85
else:
    PAG = X * 0.90
# 3. Mostrar el resultado
print(f"El pago total por los {X} lápices es de: {PAG:.2f}")
print("Fin del programa ")
print("Programa realizado por Alejandra Rivera Valenzuela")
```
