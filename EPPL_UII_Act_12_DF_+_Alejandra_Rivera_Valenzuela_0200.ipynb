```python
# para cuando es verdadero
# calculando cambiuo usando if else
# Calculando cambio usando if else
# Datos de entrada: CD (cantidad de dinero), X (cantidad de lapices), precio_unitario (precio por lapiz)

X = float(input("Ingrese la cantidad de lapices: "))
precio_unitario = float(input("Ingrese el precio unitario del lapiz: "))
if X >= 1000:
    CP = X * precio_unitario * 0.90
CD = float(input("Ingrese la cantidad de dinero: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El costo total de los lapices con descuento es: ", CP)
    print("El cambio es: ", Cambio)
    print("Fin del programa ")
    print("Programa realizado por Alejandra Rivera Valenzuela NL = 45")
```
